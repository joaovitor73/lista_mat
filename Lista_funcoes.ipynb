**Bibliotecas**

```python
#Biblioteca para manipulação de funções
import numpy as np
#Biblioteca para criação de gráficos
import matplotlib.pyplot as plt
#Instalando módulo de estilização de gráficos
!pip install mplcyberpunk
#Biblioteca para estilização de gráficos
import mplcyberpunk
#biblioteca matematica
import math
```

**Gerando x**

```python
#Gerando uma lista com valores do -5 ao 5(Domínio do estudo)
x_float = np.linspace(-5, 5, 11)
#Tranformando os valores para inteiro
x_int = np.round(x_float).astype(int)
#plotando
x = x_int
x
```

**Estilizando o gráfico**

```python
#Definindo o estilo do gráfico
plt.style.use("cyberpunk")
```

**Função para gerar gráficos**

```python
def gerar_grafico(y, legenda):
  #Criando o gráfico passando os valores de x e y com uma legenda
  plt.plot(x, y)

  #Colocando título no gráfico
  plt.title(legenda)

  #Títulos para os eixos
  plt.xlabel("Eixo X")
  plt.ylabel("Eixo Y")

  # Adicionar o efeito cyberpunk
  mplcyberpunk.add_glow_effects()

  # Adicionar manualmente as retas dos eixos X e Y
  plt.axhline(0, color='red', linestyle='--', linewidth=0.5)
  plt.axvline(0, color='red', linestyle='--', linewidth=0.5)

  # Exibir o gráfico
  plt.show()
```

**Função do primeiro grau**

```python
#Gerando a imagem
y = ((2*x)+ 3)
#plotando a imagem
gerar_grafico(y, "Gráfico Função linear")
```

**Função polimonial**

```python
#Gerando a imagem
y = ((2*(x**3))-(5*(x**2))+(3*x)-1)
#plotando a imagem
gerar_grafico(y, "Gráfico Função polimonial")
```

**Função quadrática**

```python
y = (2*(x**2)-(3*x)+1)

gerar_grafico(y, "Gráfico Função quadrática")
```

**Função modular**

```python
y = abs(x-4)

gerar_grafico(y, "Gráfico Função modula")
```

**Função exponecial**

```python
y = np.exp(x)

gerar_grafico(y, "Gráfico Função exponecial")
```

**Função logaritimica**

```python
y = np.log(x)

gerar_grafico(y, "Gráfico Função logaritmica")
```

**Função Trigonométrica**

```python
y = np.sin(x)

gerar_grafico(y, "Gráfico da função seno")
```

```python
y = np.cos(x)

gerar_grafico(y, "Gráfico da função cosseno")
```
